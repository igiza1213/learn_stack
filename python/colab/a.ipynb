```python
import numpy as np
import cv2
import random

img = np.zeros((512, 512, 3), np.uint8)
drawing = False


def write_circle(x, y):
    cv2.circle(img, (x, y), 10, tuple(random.choice(range(256)) for _ in range(3)), -1)


def draw_circle(event, x, y, flags, param):
    global drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        write_circle(x, y)
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            write_circle(x, y)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        write_circle(x, y)


cv2.namedWindow('sketch')
cv2.setMouseCallback('sketch', draw_circle)

while True:
    cv2.imshow('sketch', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
img = np.zeros((500, 600, 3), np.uint8)

cv2.imshow('image', img)



def onChange(pos):
    r = cv2.getTrackbarPos("R", "image")
    g = cv2.getTrackbarPos("G", "image")
    b = cv2.getTrackbarPos("B", "image")
    img[:, :] = [b,g,r]
    cv2.imshow('image', img)



cv2.createTrackbar("R", "image", 0, 255, onChange)
cv2.createTrackbar("G", "image", 0, 255, onChange)
cv2.createTrackbar("B", "image", 0, 255, onChange)


cv2.waitKey(0)

cv2.destroyAllWindows()
```
